### d) O que você notou em relação aos gráficos produzidos nos itens anteriores? Forneça uma análise quanto ao erro cometido.

Nota-se, a partir da análise dos gráficos obtidos, que o erro do método de Euler diminui conforme reduzimos o valor do passo $h$.
Isso se deve ao fato de que o método de Euler se utiliza da aproximação da derivada $x'(t_0) ≈ \frac{x(t)- x(t_0)}{t- t_0}$. Nessa aproximação, quanto mais próximo $t_0$ for de $t$, melhor a aproximação. 

Logo, como o passo do método de Euler é dado justamente por $t- t_0$, isto é, pela diferença entre $t$ e $t_0$, faz sentido afirmar que, quanto menor o passo, melhor o resultado! Pelo menos teoricamente (isso não acontece quando o método diverge).
