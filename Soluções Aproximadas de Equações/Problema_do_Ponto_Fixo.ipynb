## Exercício 1

Considere a função $f(x)=x²+x-1$.




###  a) e b) Faça um esboço do problema de ponto fixo desta função e determine seus pontos fixos algebricamente.

Primeiro, vamos plotar apenas o gráfico de $f(x)=x²+x-1$.

```python
# Definindo a função f(x)
def f(x):
    return x**2 + x - 1

# Criando um array de valores de x no intervalo [-2, 2]
x = np.linspace(-2, 2, 100)

# Calculando os valores de f(x) para cada valor de x
y = f(x)

# Plotando o gráfico da função f(x)
plt.plot(x, y)

# Definindo o título do gráfico e dos eixos
plt.title('Gráfico da função f(x) = x² + x - 1')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.show()
```

Agora, para resolver o problema do ponto fixo, vamos definir a função identidade h(x)= x. Então, vamos plotá-la junto ao gráfico anterior

```python
# Definindo a função identidade g(x) = x
def h(x):
    return x

# Valores do eixo y para ambas as funções
y1 = f(x)
y2 = h(x)

# Plotando o gráfico das funções f(x) e h(x)
plt.plot(x, y1, label='f(x)')
plt.plot(x, y2, label='h(x)')

# Definindo o título do gráfico e dos eixos
plt.title('Função f(x) e indentidade plotadas juntas')
plt.xlabel('x')
plt.ylabel('y')
```

Logo, pelo gráfico acima, já podemos concluir que haverá dois pontos fixos para o problema.

Para achar os pontos fixos, basta encontrar os pontos em que f(x)= h(x)!

Ou seja, basta calcular $x²+ x -1 = x →	 x² = 1 →	 x= ±1$. Plotando esses pontos fixos no gráfico, temos:

```python
# Plotando o gráfico das funções f(x) e h(x)
plt.plot(x, y1, label='f(x)')
plt.plot(x, y2, label='h(x)')

# Definindo o título do gráfico e dos eixos
plt.title('Função f(x) e indentidade com pontos fixos')
plt.xlabel('x')
plt.ylabel('y')
# Encontra os pontos de interseção entre f(x) e h(x)
points = [-1, 1]

print(points)
# Adiciona marcadores para os pontos de interseção
for p in points:
    plt.plot(p, p, marker='o', color='red')

# Define a legenda do gráfico
plt.legend()

# Exibe o gráfico
plt.show()
```

## Exercício 2

Considere a função $g(x)=(-2x²+x+3)^{1/4}$ . 


### a) Utilize o método iterativo do ponto fixo para estimar o ponto fixo da função.

Nesse exercício, diferentemente do anterior- em que pudemos facilmente encontrar os pontos fixos apenas igualando a função f(x) com a função identidade h(x)

- é necessário realizar o método da iteração do ponto fixo, que consiste em escolher um valor inicial para x e, em seguida, repetidamente aplicar a função até que a solução converge para o ponto fixo.

Escolhendo um valor inicial de $x = 0.5$, temos:

$g(0.5) = (-2(0.5)² + 0.5 +3)^{1/4} ≈ 1.156$

$g(1.156) = (-2(1.156)² + 1.156 +3)^{1/4} ≈ 1.103$


$g(1.103) = (-2(1.103)² + 1.103 +3)^{1/4} ≈ 1.103$


$g(1.136) = (-2(1.136)² + 1.136 +3)^{1/4} ≈ 1.117$


$g(1.117) = (-2(1.117)² + 1.117 + 3)^{1/4} ≈ 1.128$


$g(1.128) = (-2(1.128)² + 1.128 + 3)^{1/4} ≈ 1.128$


$g(1.121) = (-2(1.121)² + 1.121 + 3)^{1/4} ≈ 1.126$


$g(1.126) = (-2(1.126)² + 1.126 + 3)^{1/4} ≈ 1.122$

Continuando esse processo, podemos ver que a solução converge para um valor em torno de 1.124. Portanto, a resposta aproximada para o ponto fixo da função é x ≈ 1.124.

Ao invés de fazer "na mão", podemos desenvolver um pequeno código que também chega a um resultado semelhante:

```python
def ponto_fixo(f, x, eps):
    """
    Encontra uma raiz da função f usando o método iterativo do ponto fix,
    com chute inicial x e critério de parada eps.
    
    Retorna a raiz encontrada e o número de iterações.
    """
    i = 0
    while True:
        x_next = f(x) # o x seguinte será o valor de g(x) medido anteriormente
        if abs(x_next - x) < eps: # o código vai parar quando os x obtidos praticamente pararem de variar
            break
        x = x_next

    print(f"A solução é x = {x_next}. E o número de iterações foi i = {i}")
```

```python
def g(x): # definindo a função g(x)
    return (-2*x**2 + x + 3)**(1/4)

x = 0.5 # chute inicial
tol = 1e-6 # critério de parada
```

```python
ponto_fixo(g, x, tol)
```

###  b) Faça um esboço geométrico do método para este problema específico. 


Primeiro, vamos plotar apenas o gráfico de g(x) =$(-2x²+x+3)^{1/4}$.

```python
# Definindo a função g(x)
def g(x):
    return (-2*(x**2) + x +3)**(1/4)

# Criando um array de valores de x no intervalo [-2, 2]
xg = np.linspace(-1, 1.5, 1000)

# Calculando os valores de g(x) para cada valor de x
yg = g(xg)

# Plotando o gráfico da função g(x)

plt.plot(xg, yg)

# Definindo o título do gráfico e dos eixos
plt.title('Gráfico da função g(x) =$(-2x²+x+3)^{1/4}$')
plt.xlabel('x')
plt.ylabel('g(x)')

plt.show()
```

Agora, para resolver o problema do ponto fixo, vamos definir a função identidade h(x)= x. Então, vamos plotá-la junto ao gráfico anterior

```python
# Definindo a função identidade g(x) = x
def h(x):
    return x

# Valores do eixo y para ambas as funções
y3 = g(xg)
y4 = h(xg)

# Plotando o gráfico das funções f(x) e h(x)
plt.plot(xg, y3, label='g(x)')
plt.plot(xg, y4, label='h(x)')

# Definindo o título do gráfico e dos eixos
plt.title('Função g(x) e indentidade plotadas juntas')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(bottom=0)
```

Segundo o resultado que obtivemos na questão anterior, o ponto fixo desse problema é cerca de 1.124
Plotando isso no gráfico temos:

```python
# Plotando o gráfico das funções g(x) e h(x)
plt.plot(xg, y3, label='f(x)')
plt.plot(xg, y4, label='h(x)')

# Definindo o título do gráfico e dos eixos
plt.title('Função g(x) e indentidade com ponto fixo')
plt.xlabel('xg')
plt.ylabel('yg')
plt.ylim(bottom=0)
# Ponto de interseção entre g(x) e h(x)
points2 = [1.124]

print(points2)
# Adiciona marcadores para os pontos de interseção
for p in points2:
    plt.plot(p, p, marker='o', color='red')

# Define a legenda do gráfico
plt.legend()
plt.ylim(bottom=0)
# Exibe o gráfico
plt.show()
```
