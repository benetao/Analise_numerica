Nota-se, a partir da anlise dos grficos obtidos, que o erro do mtodo de Euler diminui conforme reduzimos o valor do passo $h$.
Isso se deve ao fato de que o mtodo de Euler se utiliza da aproximao da derivada $x'(t_0)  \frac{x(t)- x(t_0)}{t- t_0}$. Nessa aproximao, quanto mais prximo $t_0$ for de $t$, melhor a aproximao. 

Logo, como o passo do mtodo de Euler  dado justamente por $t- t_0$, isto , pela diferena entre $t$ e $t_0$, faz sentido afirmar que, quanto menor o passo, melhor o resultado! Pelo menos teoricamente (isso no acontece quando o mtodo diverge).
