##  a)  Para cada modelo esboce o grfico das solues analticas, juntamente com as solues numricas. Nesse problema, considere condio inicial  $x(0)=0.5$, espaamento $h=1$ e constante de crescimento $=0.1$.

### 1- Malthus

```python
def runge_kutta_4(f, y0, t0, tf, dt):
    """
    Resolve uma equao diferencial ordinria pelo mtodo de Runge-Kutta de ordem 4.

    Parmetros:
    f: funo que descreve a equao diferencial (deve ter a assinatura f(t, y)).
    y0: valor inicial da soluo.
    t0: tempo inicial.
    tf: tempo final.
    dt: tamanho do passo de tempo.

    Retorno:
    Um array com a soluo da equao diferencial nos tempos especificados.
    """

    # Inicializa o vetor de tempo e de soluo
    t = np.arange(t0, tf + dt, dt)
    y = np.zeros(len(t))
    y[0] = y0

    # Implementa o mtodo de Runge-Kutta de ordem 4
    for i in range(len(t) - 1):
        k1 = dt * f(t[i], y[i])
        k2 = dt * f(t[i] + dt/2, y[i] + k1/2)
        k3 = dt * f(t[i] + dt/2, y[i] + k2/2)
        k4 = dt * f(t[i] + dt, y[i] + k3)
        y[i+1] = y[i] + 1/6 * (k1 + 2*k2 + 2*k3 + k4)

    return y
```

```python
def metodo_euler(f, y0, h, crit_parada):
    """
    Implementa mtodo de Euler para resolver a equao de Malthus.

    Args:
        f : funo que define a equao diferencial.
        x0 : valor inicial da populao.
        h : tamanho do passo de integrao.
        crit_parada : nmero de iteraes do cdigo.

    Returns:
        listas do eixo x e do eixo y para plotar o grfico
    """
    y_euler = [y0]  # Lista para armazenar as solues do eixo y
    x_euler = [0]  # Lista para armazenar valores do eixo x
    y = y0  # Valor inicial

    for i in range(crit_parada):
        y += h * f(x_euler[-1], y)  # Atualiza y usando o mtodo de Euler
        t = (i + 1) * h  # Atualiza t
        x_euler.append(t)
        y_euler.append(y)

    return x_euler, y_euler
```

```python
# Funo que define a equao diferencial de Malthus com taxa de crescimento populacional= 0.1
def equacao_malthus(t,x):
    return 0.1 * x
```

```python
# Parmetros do problema
x0 = 0.5  # Valor inicial da populao
h = 1  # Tamanho do passo
crit_parada = 100  # Nmero de iteraes
lamb = 0.1  # Taxa de crescimento populacional
t0=0 # tempo inicial
tf= 100 # tempo final

# Valores do grfico do mtodo de Euler
x_malthus_euler, y_malthus_euler = metodo_euler(equacao_malthus, x0, h, crit_parada)

# Valores do grfico do mtodo de Runge-Kutta
y_malthus_rk= runge_kutta_4(equacao_malthus, x0, t0, tf, h)

# Valores do grfico real
y_malthus_real = x0 * np.exp(lamb * np.array(x_malthus_euler))


# Plotando o grfico da soluo obtida pelo mtodo de Euler
plt.plot(x_malthus_euler, y_malthus_euler, label='Mtodo de Euler')
# Plotando o grfico da soluo obtida pelo mtodo de Runge-Kutta
plt.plot(x_malthus_euler, y_malthus_rk, label='Runge-Kutta')
# Plotando o grfico da soluo real
plt.plot(x_malthus_euler, y_malthus_real, label='Soluo Real', linestyle='dashdot')
plt.xlabel('Tempo (t)')
plt.ylabel('Populao (y)')
plt.title('Equao Diferencial de Malthus')
plt.legend()
plt.show()
```

### 2- Verhulst

```python
def verhulst(t,x):
    return 0.1 * x*(1-x)
```

```python
# Valores do grfico do mtodo de Euler
x_verhulst_euler, y_verhulst_euler = metodo_euler(verhulst, x0, h, 50)

# Valores do grfico do mtodo de Runge-Kutta
y_verhulst_rk= runge_kutta_4(verhulst, x0, 0, 50, h)

# Valores do grfico real

y_verhulst_real=  1/(1 + ((1 - x0) / x0) * np.exp(-lamb * np.array(x_verhulst_euler)))

# Plotando o grfico da soluo obtida pelo mtodo de Euler
plt.plot(x_verhulst_euler, y_verhulst_euler, label='Mtodo de Euler')
# Plotando o grfico da soluo obtida pelo mtodo de Runge-Kutta
plt.plot(x_verhulst_euler, y_verhulst_rk, label='Runge-Kutta')
# Plotando o grfico da soluo real
plt.plot(x_verhulst_euler, y_verhulst_real, label='Soluo Real', linestyle='dashdot')
plt.xlabel('Tempo (t)')
plt.ylabel('Populao (y)')
plt.title('Equao Diferencial de Verhulst')
plt.legend()
plt.show()
```

### 3- Gompertz

```python
def gompertz(t,x):
    return 0.1 * x* np.log(1/x)
```

```python
# Valores do grfico do mtodo de Euler
x_gom_euler, y_gom_euler = metodo_euler(gompertz, x0, h, 50)

# Valores do grfico do mtodo de Runge-Kutta
y_gom_rk= runge_kutta_4(gompertz, x0, 0, 50, h)

# Valores do grfico real

y_gom_real=  np.exp(-np.log(2)*np.exp(-lamb* np.array(x_gom_euler)))

# Plotando o grfico da soluo obtida pelo mtodo de Euler
plt.plot(x_gom_euler, y_gom_euler, label='Mtodo de Euler')
# Plotando o grfico da soluo obtida pelo mtodo de Runge-Kutta
plt.plot(x_gom_euler, y_gom_rk, label='Runge-Kutta')
# Plotando o grfico da soluo real
plt.plot(x_gom_euler, y_gom_real, label='Soluo Real', linestyle='dashdot')
plt.xlabel('Tempo (t)')
plt.ylabel('Populao (y)')
plt.title('Equao Diferencial de Gompertz')
plt.legend()
plt.show()
```

## b) Compare os dois mtodos numricos em relao a soluo analtica de cada problema. O que pode ser dito sobre a eficcia das aproximaes?

Comparando as solues dadas por ambos os mtodos com a soluo analtica, percebe-se que o mtodo de Runge-Kutta gerou resultados mais prximos que o de Euler. Isso fica claro sobretudo na soluo da equao de Malthus, em que o Mtodo de Euler gera uma soluo muito menos prxima da analtica que o mtodo de Runge-Kutta, como podemos ver no grfico plotado no item a) e tambm pelo erro que pode ser calculado por:

```python
erro_malthus_euler = np.linalg.norm(y_malthus_real - y_malthus_euler)
print('Erro do mtodo de Euler em relao  soluo analtica da equao de Malthus:', erro_malthus_euler)
erro_malthus_rk = np.linalg.norm(y_malthus_real - y_malthus_rk)
print('Erro do mtodo de Runge Kutta em relao  soluo analtica da equao de Malthus:', erro_malthus_rk)
```

J na soluo da equao diferencial de Verhulst, ambos os mtodos resultaram em solues prximas da real, mas, pelo grfico e pelo erro, podemos concluir que o resultado pelo mtodo de Runge-Kutta foi mais preciso:

```python
error_verhulst_euler = np.linalg.norm(y_verhulst_real - y_verhulst_euler)
print('Erro do mtodo de Euler em relao  soluo analtica da equao de Verhulst:', error_verhulst_euler)
error_verhulst_rk = np.linalg.norm(y_verhulst_real - y_verhulst_rk)
print('Erro do mtodo de Runge Kutta em relao  soluo analtica da equao de Verhulst:', error_verhulst_rk)
```

Por fim, a soluo da equao de Gompertz foi melhor aproximada pelo mtodo de Runge-Kutta, que, assim como na equao de Malthus e de Verhulst, praticamente sobreps a soluo analtica no grfico e apresenta um erro muito pequeno:

```python
error_gom_euler = np.linalg.norm(y_gom_real - y_gom_euler)
print('Erro do mtodo de Euler em relao  soluo analtica da equao de Gompertz:', error_gom_euler)
error_gom_rk = np.linalg.norm(y_gom_real - y_gom_rk)
print('Erro do mtodo de Runge Kutta em relao  soluo analtica da equao de Gompertz:', error_gom_rk)
```

### Observao: como funciona o mtodo `linalg.norm`

O linalg.norm, da biblioteca numpy, calcula a norma da matriz inserida no argumento. No caso, colocamos como argumento uma matriz linha referente  diferena entre os valores de y dados pelo mtodo de Euler ou de Range-Kutta e pelo soluo analtica. Como no adicionamos nenhum argumento que especificasse qual tipo de norma queremos, o linalg.norm calcula a norma de Frobenius, que  basicamente a raiz quadrada da soma dos quadrados dos elementos da matriz. Dessa forma, quanto maior o resultado dessa norma, maior a diferena entre o resultado dado pelo mtodo em relao  soluo real e, portanto, maior o erro do mtodo usado.
